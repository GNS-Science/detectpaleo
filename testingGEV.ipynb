```python
# A test workpout for generalized extreme distribution

from scipy.stats import genextreme as gev
import matplotlib.pyplot as plt
import numpy as np
```

```python
## nsampling = 100  # idealy 10,000
nsampling =100
maxslp = 5  # again a rough estimate
sampledslip = np.linspace(0.0,maxslp,nsampling) 

# shape, scale and location
p = gev.pdf(sampledslip, -0.56, loc=0.29, scale=0.24)
plt.figure(figsize=(4, 4))
plt.plot(sampledslip, p, '-');
```

```python
# shape, scale and location
p = gev.pdf(sampledslip, -0.37, loc=0.37, scale=0.37)
plt.figure(figsize=(4, 4))
plt.plot(sampledslip, p, '-');
```
