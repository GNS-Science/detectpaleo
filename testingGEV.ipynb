```python
# A test workpout for generalized extreme distribution

from scipy.stats import genextreme as gev
import matplotlib.pyplot as plt
import numpy as np
```

```python
nsampling = 100  # idealy 10,000
maxslp = 5   # again a rough estimate
sampledslip = np.linspace(0.001,maxslp,nsampling) 
p = gev.pdf(sampledslip, 0.24, loc=0.56, scale=0.29)

plt.plot(sampledslip, p, 'o-')
```
